# First steps com dados BTC

```python
# Import PySpark and create a SparkSession
from pyspark.sql import SparkSession

spark = SparkSession.builder.appName("ImportData").getOrCreate()
```

```python
spark
```

# Ler o ficheiro BTC (Parquet)


```python
# Ler dados BTC em formato Parquet (dados reais da Binance)
df_btc = spark.read.parquet("./../data/btc_04h_usdt_binance.parquet")
```

```python
df_btc.show(truncate=5)
```

# Filter operations
* Filtrar por preco close e volume


```python
# Filtrar dias com close acima de 50000
df_btc.filter("close>=50000").show(truncate=5)
```

```python
# Filtrar dias com close acima de 50000
df_btc.filter("close>=50000").show(truncate=5)
```

```python
# Filtrar dias com close acima de 50000
df_btc.filter("close>=50000").show(truncate=5)
```

```python
# Filtrar usando sintaxe de objeto
df_btc.filter(df_btc.close > 50000).show(truncate=5)
```

```python
# Filtrar usando sintaxe de dicionario
df_btc.filter(df_btc['close'] > 50000).show(truncate=3)
```

Podemos usar:
* | or
* & and
* ~ (negate)

```python
# Filtrar close > 50000 AND volume < 1000000000
df_btc.filter((df_btc['close'] > 50000) & (~(df_btc['volume'] > 10000000000))).show(truncate=4)
```

```python
# Filtrar close > 50000 AND volume > 10000000000
df_btc.filter((df_btc['close'] > 50000) & (df_btc['volume'] > 10000000000)).show(truncate=4)
```
